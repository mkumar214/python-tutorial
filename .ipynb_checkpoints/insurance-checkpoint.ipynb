```python
#import all required library
#numpy : numpy for matries , multidimentional arrary and mathematical calculation
import numpy as np
#pandas: for data manipulation and analysis
import pandas as pd
#matplotlib : plottin library
import matplotlib.pyplot as plt
#seaborn :better data visualisation
import seaborn as sns
```

```python
#Five piont summary include
#Minimum
#Max
#1st Quartile
#2nd Quartile (Median)
#3rd Quartile 


#minimum
insurance[['age','bmi','children','charges']].min()
```

```python
sns.boxplot(x=insurance[['age','bmi','children','charges']])  # boxplot for outlier
```

```python
#outlier compute by z-score
from scipy import stats
z_outlier=np.abs(stats.zscore(insurance[['age','bmi','charges']]))
print(z_outlier)
```

```python
#fltering on threshold vaule
threshold=3
np.where(z_outlier > threshold)
```
